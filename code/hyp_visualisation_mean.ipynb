```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# hypothesis 1

data = pd.read_csv('../results/df_hypothesis1.csv', index_col = 'Unnamed: 0')
g = sns.catplot(data = data,
                kind = 'strip', # other possible keywords: 'swarm', 'violin', 'strip', 'box', 'boxen', 'point', 'bar'
                x = 'condition',
                y = 'minimum',
                alpha = 0.4, # transparency - needs to be commented out for some kinds of plots (e.g. 'box')
                palette = ['mediumslateblue', 'orangered'],
                # col = 'condition', # split the data into subplots next to each other - try 'stimulus type' (and set x to 'condition')
                hue = 'condition', # some plot kinds allow for a clearer seperation between hues, e.g. 'point'
                legend = False
               ) 
sns.pointplot(data = data,
              x = 'condition',
              y = 'minimum',
              ax = g.ax,
              palette = ['mediumslateblue', 'orangered'],
              linestyles= [' ', ' '],
              #legend = False,
              # col = 'condition',
              hue = 'condition', # some plot kinds allow for a clearer seperation between hues, e.g. 'point'
           );

g.ax.set_xlabel('Condition')
g.ax.set_ylabel('Minimum saccade rate [Hz]')
g.ax.set_title('Minimum saccade rates by condition')
g.ax.legend(loc='upper right')
g.fig.set_size_inches(6.5, 5)
plt.tight_layout()
g.figure.savefig('../results/visualisations/visualization_of_means_hyp1_condition.png')
```
