# causal rate analysis 

```python
def causal_rate(move_onset, lock_window_start, lock_window_end, n_trials, alpha):
    """
     analyse rate in causal time window

     input:    move_onset  - movement onset times
               lock_window_start  - window before lock
               lock_window_end  - window after lock
               n_trials      - number of trials

     output:   rate    - movement rate
               scale   - time axis

    12.12.2005 by Martin Rolfs
    21.06.2021 translated to python by Clara Kuper
    """
    scale = np.arange(lock_window_start, lock_window_end, 1)
    # check how many trials these values came from
    if type(n_trials) == int:
        n_trials = np.linspace(n_trials, n_trials, len(scale))
    elif len(n_trials) != len(scale):
        raise ValueError('n_trials must have the same as the length of lock_window_start:lock_window_end!'
                         f'But has length {len(n_trials)} instead of {len(scale)}')
    # alpha defines how much the distribution is shifted
    alpha = alpha
    # define empty arrays for scale and rate
    rate = []
    raw_rate = []

    # loop through all time windows
    for idx, t in enumerate(scale):
        # compute tau
        # here is a filter for all events BEFORE time point t
        tau = t - move_onset + 1 / alpha
        # filter tau as event 0/1
        tau = tau[tau > 0]
        # get the number of saccades in a given window
        causal = alpha ** 2 * tau * np.exp(-alpha * tau)
        # save the rate
        rate.append(sum(causal) * 1000 / n_trials[idx])
        raw_rate.append(sum(causal) * 1000)
    return np.array(raw_rate), np.array(rate), scale
```

## define functions

```python
def get_indices_by_value(scale, start_value, end_value):
    start_index = np.where(scale == start_value)[0]
    end_index = np.where(scale == end_value)[0]
    assert len(start_index) == len(end_index) == 1
    return start_index[0], end_index[0]
```

```python
def get_mean_between_indices(rate, start_index, end_index):
    rate_window = rate[start_index : end_index]
    #print(rate_window)
    return np.mean(rate_window)
```

```python
def get_mean_between_values(scale, rate, start_value, end_value):
    start_index, end_index = get_indices_by_value(scale, start_value, end_value)
    return get_mean_between_indices(rate, start_index, end_index)
```

```python
def get_minimum_between_indices(rate, start_index, end_index):
    rate_window = rate[start_index : end_index]
    #print(rate_window)
    return np.min(rate_window)
```

```python
def get_minimum_between_values(scale, rate, start_value, end_value):
    start_index, end_index = get_indices_by_value(scale, start_value, end_value)
    return get_minimum_between_indices(rate, start_index, end_index)
```

```python
def get_maximum_between_indices(rate, start_index, end_index):
    rate_window = rate[start_index : end_index]
    #print(rate_window)
    return np.max(rate_window)
```

```python
def get_maximum_between_values(scale, rate, start_value, end_value):
    start_index, end_index = get_indices_by_value(scale, start_value, end_value)
    return get_maximum_between_indices(rate, start_index, end_index)
```

```python
# gaze only per participant 

base_all_g = [] #relevant_base
control_all_g = [] # irrelevant_base
relevant_all_g = []
irrelevant_all_g = []

df_hyp1 = pd.DataFrame(columns=['participant', 
                                'condition', 
                                'stimulus type', 
                                'mean', 
                                'minimum']) 

df_hyp2 = pd.DataFrame(columns=['participant', 
                                'condition', 
                                'stimulus type', 
                                'mean', 
                                'minimum'])

df_hyp4 = pd.DataFrame(columns=['participant', 
                                'condition', 
                                'stimulus type', 
                                'maximum'])

for p in ppts:
    data_g = pd.read_csv(f'../data/participant_{p}_gaze_saccades.csv')
    print(len(data_g.groupby(['SESSION', 'BLOCK', 'TRIAL'])))
    data_g['ONSET_RELATIVE_RELEVANT'] = data_g['onset'] - data_g['TRIAL_CLOSEST_RELEVANT_TIME']
    data_g['ONSET_RELATIVE_IRRELEVANT'] = data_g['onset'] - data_g['TRIAL_CLOSEST_IRRELEVANT_TIME']
    baseline_data_g = data_g[(data_g['CLOSEST_RELEVANT_SHOWN'] == 0) & (data_g['CLOSEST_IRRELEVANT_SHOWN'] == 0)]
    relevant_data_g = data_g[(data_g['CLOSEST_RELEVANT_SHOWN'] == 1) & (data_g['CLOSEST_IRRELEVANT_SHOWN'] == 0)]
    irrelevant_data_g = data_g[(data_g['CLOSEST_RELEVANT_SHOWN'] == 0) & (data_g['CLOSEST_IRRELEVANT_SHOWN'] == 1)]

    # rate for baseline condition
    base_rate_raw_g, base_rate_g, scale_g = causal_rate(baseline_data_g["ONSET_RELATIVE_RELEVANT"].values * 1000, 
                                                        start, 
                                                        end, 
                                                        len(baseline_data_g.groupby(['TRIAL', "SESSION", "BLOCK"]).describe()), 
                                                        alpha)
    # rate for baseline condition II
    control_rate_raw_g, control_rate_g, scale_g = causal_rate(baseline_data_g["ONSET_RELATIVE_IRRELEVANT"].values * 1000, 
                                                              start, 
                                                              end, 
                                                              len(baseline_data_g.groupby(['TRIAL', "SESSION", "BLOCK"]).describe()), 
                                                              alpha)
    # rate for relevant condition
    relevant_rate_raw_g, relevant_rate_g, scale_g = causal_rate(relevant_data_g["ONSET_RELATIVE_RELEVANT"].values * 1000, 
                                                                start, 
                                                                end, 
                                                                len(relevant_data_g.groupby(['TRIAL', "SESSION", "BLOCK"]).describe()), 
                                                                alpha)
    # rate for irrrelevant condition 
    irrelevant_rate_raw_g, irrelevant_rate_g, scale_g = causal_rate(irrelevant_data_g["ONSET_RELATIVE_IRRELEVANT"].values * 1000, 
                                                                    start, 
                                                                    end, 
                                                                    len(irrelevant_data_g.groupby(['TRIAL', "SESSION", "BLOCK"]).describe()), 
                                                                    alpha)

    df_hyp1 = df_hyp1.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'irrelevant', 
                              'mean':get_mean_between_values(scale_g, irrelevant_rate_g, start_value_hyp1, end_value_hyp1), 
                              'minimum': get_minimum_between_values(scale_g, irrelevant_rate_g, start_value_min, end_value_min)}, 
                             ignore_index = True)
    df_hyp1 = df_hyp1.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'relevant', 
                              'mean':get_mean_between_values(scale_g, relevant_rate_g, start_value_hyp1, end_value_hyp1),
                              'minimum': get_minimum_between_values(scale_g, relevant_rate_g, start_value_min, end_value_min)}, 
                             ignore_index = True)
    df_hyp1 = df_hyp1.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'baseline', 
                              'mean':get_mean_between_values(scale_g, base_rate_g, start_value_hyp1, end_value_hyp1),
                              'minimum': get_minimum_between_values(scale_g, base_rate_g, start_value_min, end_value_min)}, 
                             ignore_index = True)
    
    df_hyp2 = df_hyp2.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'irrelevant', 
                              'mean':get_mean_between_values(scale_g, irrelevant_rate_g, start_value_hyp2, end_value_hyp2)}, 
                             ignore_index = True)
    df_hyp2 = df_hyp2.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'relevant', 
                              'mean':get_mean_between_values(scale_g, relevant_rate_g, start_value_hyp2, end_value_hyp2)}, 
                             ignore_index = True)
    df_hyp2 = df_hyp2.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'baseline', 
                              'mean':get_mean_between_values(scale_g, base_rate_g, start_value_hyp2, end_value_hyp2)}, 
                             ignore_index = True)
    
    df_hyp4 = df_hyp4.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'irrelevant', 
                              'maximum': get_maximum_between_values(scale_g, irrelevant_rate_g, start_value_max, end_value_max)}, 
                             ignore_index = True)
    df_hyp4 = df_hyp4.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'relevant', 
                              'maximum': get_maximum_between_values(scale_g, relevant_rate_g, start_value_min, end_value_min)}, 
                             ignore_index = True)
    df_hyp4 = df_hyp4.append({'participant':p, 
                              'condition': 'gaze', 
                              'stimulus type':'baseline', 
                              'maximum': get_maximum_between_values(scale_g, base_rate_g, start_value_min, end_value_min)}, 
                             ignore_index = True)
    
    base_all_g.append(base_rate_g)
    control_all_g.append(control_rate_g)
    relevant_all_g.append(relevant_rate_g)
    irrelevant_all_g.append(irrelevant_rate_g)
    
    # Create participant directory if it does not exist
    participant_dir = f'../data/{p}'
    if not os.path.exists(participant_dir):
        os.makedirs(participant_dir)
    
    # Define the output path for the plot
    output_path = os.path.join(participant_dir, f'{p}_plot_gaze.png')

    plt.plot(scale_g, base_rate_g, 
             color = 'black', 
             linestyle = '--', 
             label = 'baseline')
    plt.plot(scale_g, relevant_rate_g, 
             color = 'mediumslateblue', 
             linestyle = '-', 
             label = 'relevant stimulus')
    plt.plot(scale_g, irrelevant_rate_g, 
             color = 'skyblue', 
             linestyle = '-', 
             label = 'irrelevant stimulus')
    
    plt.ylim(0, 9)
    plt.xlabel('Time from stimulus onset [ms]')
    plt.ylabel('Saccade rate [Hz]')
    plt.legend()
    plt.axvline(x=0, color='k', linestyle='-', alpha = 0.5)
    plt.savefig(output_path)
    plt.show()
```

## define functions

```python
def get_rates_for_normalization(all_rates_to_normalize, all_rates_to_normalize_by, participant_id):
    rate_to_normalize = all_rates_to_normalize[participant_id]
    rates_to_normalize_by = np.delete(all_rates_to_normalize_by, participant_id, axis=0)
    
    return rate_to_normalize, rates_to_normalize_by
```

```python
def get_normalized_rate(rate_to_normalize, rates_to_normalize_by):
    normalized_rate = rate_to_normalize / np.mean(rates_to_normalize_by, axis=0)
    
    return normalized_rate 
```

# minimum rates, maximum rates, latencies

## define functions

```python
def get_latency(min_value_list, rate_list, scale):
    index_list = []
    for idx, val in enumerate(min_value_list):
        minimum_index = np.where(rate_list[idx] == val)[0][0]
        index_list.append(scale[minimum_index])
    return index_list
```

```python
# minimum saccade rate 0 - 200 ms (inhibition, dip)

min_relevant_g = [min(rate[100:300]) for rate in normalized_relevant_g]
latency_min_relevant_g = get_latency(min_relevant_g, normalized_relevant_g, scale_g)

min_irrelevant_g = [min(rate[100:300]) for rate in normalized_irrelevant_g]
latency_min_irrelevant_g = get_latency(min_irrelevant_g, normalized_irrelevant_g, scale_g)

min_relevant_m = [min(rate[100:300]) for rate in normalized_relevant_m]
latency_min_relevant_m = get_latency(min_relevant_m, normalized_relevant_m, scale_m)

min_irrelevant_m = [min(rate[100:300]) for rate in normalized_irrelevant_m]
latency_min_irrelevant_m = get_latency(min_irrelevant_m, normalized_irrelevant_m, scale_m)
```

```python
# maximum saccade rate 100 - 300 ms (recovery, peak)

max_relevant_g = [max(rate[200:400]) for rate in normalized_relevant_g]
latency_max_relevant_g = get_latency(max_relevant_g, normalized_relevant_g, scale_g)

max_irrelevant_g = [max(rate[200:400]) for rate in normalized_irrelevant_g]
latency_max_irrelevant_g = get_latency(max_irrelevant_g, normalized_irrelevant_g, scale_g)

max_relevant_m = [max(rate[200:400]) for rate in normalized_relevant_m]
latency_max_relevant_m = get_latency(max_relevant_m, normalized_relevant_m, scale_m)

max_irrelevant_m = [max(rate[200:400]) for rate in normalized_irrelevant_m]
latency_max_irrelevant_m = get_latency(max_irrelevant_m, normalized_irrelevant_m, scale_m)
```

```python
# List of participants
participants = list(range(1, 21))
```

```python
# Create lists for dataframe columns
participants_col = participants * 4
condition_col = ['gaze'] * 40 + ['move'] * 40
stimulus_type_col = ['relevant'] * 20 + ['irrelevant'] * 20 + ['relevant'] * 20 + ['irrelevant'] * 20
#minimum
normalized_minimum_rate_col = min_relevant_g + min_irrelevant_g + min_relevant_m + min_irrelevant_m
magnitude_max_inhib_col = [1 - x for x in min_relevant_g] + [1 - x for x in min_irrelevant_g] + [1 - x for x in min_relevant_m] + [1 - x for x in min_irrelevant_m]
latency_minimum_rate_col = latency_min_relevant_g + latency_min_irrelevant_g + latency_min_relevant_m + latency_min_irrelevant_m
#maximum
normalized_maximum_rate_col = max_relevant_g + max_irrelevant_g + max_relevant_m + max_irrelevant_m
magnitude_max_peak_col = [x - 1 for x in max_relevant_g ] + [x - 1 for x in max_irrelevant_g] + [ x - 1 for x in max_relevant_m] + [x - 1 for x in max_irrelevant_m]
latency_maximum_rate_col = latency_max_relevant_g + latency_max_irrelevant_g + latency_max_relevant_m + latency_max_irrelevant_m
```
