# merge dataframes

```python
#rename dataframes
df1 = data_thousandhz
df2 = data_onehundredhz
df3 = data_fiftyhz
```

```python
#add column "Sampling Rate"
df1["Sampling Rate"] = 1000
df2["Sampling Rate"] = 100
df3["Sampling Rate"] = 50
```

```python
#merge all dataframes
df_13 = pd.concat([df1, df3], axis=0)
df_13.reset_index(drop=True)
```

```python
df_13.groupby("Sampling Rate").describe()
```

```python
filtered_df = df_13[df_13['label'].isin(['SACC', 'ISAC'])]

filtered_df = filtered_df.copy()

# Combine 'SACC' and 'ISAC' into a single category 'saccadic events'
filtered_df.loc[:, 'saccadic_events'] = filtered_df['label'].replace({'SACC': 'saccadic events', 'ISAC': 'saccadic events'})

# Group by 'saccadic_events' and 'Sampling Rate', then count the occurrences
grouped_df = filtered_df.groupby(['saccadic_events', 'Sampling Rate']).size().reset_index(name='count')
```

```python
# Group by 'Sampling Rate' to count the occurrences of saccadic events
counts = filtered_df.groupby('Sampling Rate')['saccadic_events'].count()

# Calculate the percentage of 50 Hz events compared to 1000 Hz
percentage_50hz_vs_1000hz = (counts[50] / counts[1000]) * 100

print(percentage_50hz_vs_1000hz)
```

## visualize merged df

```python
# Plot the combined data

sns.set_palette("Paired")

frequ_plot = sns.barplot(data = grouped_df,
                         x = 'saccadic_events',
                         y = 'count',
                         hue = 'Sampling Rate')

frequ_plot.set_xticklabels([''])  # This removes the text on the x-axis ticks
frequ_plot.set_xlabel('Saccade events detected by REMoDNaV')

new_labels = ['50 Hz', '1000 Hz']
handles, _ = frequ_plot.get_legend_handles_labels()
frequ_plot.legend(handles, new_labels, loc='upper left')

sns.despine()

frequ_plot.set_ylabel('Count')

plt.savefig('C:/Users/anton/OneDrive/Dokumente/saccadicinhibition_bachelor_thesis/results/visualisations/downsampling.png')

plt.show()
```
