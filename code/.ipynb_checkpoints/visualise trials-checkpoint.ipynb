```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
```

```python
# GAZE gaze path on screen

data_raw = pd.read_csv('../data/01/gaze/gaze_data_01_gaze.csv')
data_sacc = pd.read_csv('../data/participant_01_gaze_saccades.csv')

data_raw_trial = data_raw[(data_raw['TRIAL'] == 70) & (data_raw['SESSION'] == 2)]
data_sacc_trial = data_sacc[(data_sacc['TRIAL'] == 70) & (data_sacc['SESSION'] == 2)]

x_pos = data_raw_trial['X_POS_SCREEN']
y_pos = data_raw_trial['Y_POS_SCREEN']

start_x = data_sacc_trial['start_x']
start_y = data_sacc_trial['start_y']

end_x = data_sacc_trial['end_x']
end_y = data_sacc_trial['end_y']

trial_time = data_raw_trial['TIME'] - data_raw_trial['TIME'].iloc[0]

for x1, x2 in zip(start_x.values,end_x.values):
    plt.fill_between([x1, x2], [0,0] ,[300,300], color='skyblue', alpha = 0.5)

plt.plot(x_pos, y_pos, linestyle='dotted', color='black', label = 'eye movement path')
#plt.scatter(start_x, start_y, color = 'blue')
#plt.scatter(end_x, end_y, color = 'magenta')

plt.xlabel("x position")
plt.ylabel("y position")
plt.savefig('../results/visualisations/participant_01_trial70_gaze_pathplot')
plt.show()
```

```python
# GAZE gaze positions across time

data_raw = pd.read_csv('../data/01/gaze/gaze_data_01_gaze.csv')
data_sacc = pd.read_csv('../data/participant_01_gaze_saccades.csv')

data_raw_trial = data_raw[(data_raw['TRIAL'] == 70) & (data_raw['SESSION'] == 2)]
data_sacc_trial = data_sacc[(data_sacc['TRIAL'] == 70) & (data_sacc['SESSION'] == 2)]

x_pos = data_raw_trial['X_POS_SCREEN']
y_pos = data_raw_trial['Y_POS_SCREEN']

start_x = data_sacc_trial['start_x']
start_y = data_sacc_trial['start_y']

end_x = data_sacc_trial['end_x']
end_y = data_sacc_trial['end_y']

trial_time = data_raw_trial['TIME'] - data_raw_trial['TIME'].iloc[0]

#plt.scatter(data_sacc_70['onset'], start_y, color = 'blue')
#plt.scatter(data_sacc_70['onset'] + data_sacc_70['duration'], end_y, color = 'magenta')
#plt.scatter(data_sacc_70['onset'], start_x, color = 'green')
#plt.scatter(data_sacc_70['onset'] + data_sacc_70['duration'], end_x, color = 'orange')

plt.xlabel("Time in seconds")
plt.ylabel("Coordinates")

for x,y in zip(data_sacc_trial['onset'].values, data_sacc_trial['onset'].values + data_sacc_trial['duration'].values):
    plt.fill_between([x, y], [0,0] ,[500,500], color='skyblue', alpha = 0.5)

plt.plot(trial_time, x_pos, linestyle='--', color='black', label='X Position')
plt.plot(trial_time, y_pos, linestyle='-', color='black', label='Y Position')

plt.legend()
plt.savefig('../results/visualisations/participant_01_trial70_gaze_timeplot')
plt.show()
```
